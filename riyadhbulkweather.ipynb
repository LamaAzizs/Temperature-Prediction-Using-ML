```python
columns_to_drop = ['dt', 'timezone', 'city_name', 'lat', 'lon', 'weather_icon']
df = df.drop(columns=columns_to_drop, errors='ignore')
```

```python
numerical_cols = df.select_dtypes(include=['float', 'int']).columns
df[numerical_cols]=df[numerical_cols].fillna(df[numerical_cols].mean())
```

```python
df['dt_iso'] = pd.to_datetime(df['dt_iso'], format='%Y-%m-%d %H:%M:%S +0000 UTC', errors='coerce')


df['dt_iso'] = df['dt_iso'].dt.strftime('%Y-%m-%d')


df['dt_iso'] = pd.to_datetime(df['dt_iso'])


df['month'] = df['dt_iso'].dt.month


df['season'] = df['month'].apply(
    lambda x: 'Winter' if x in [12, 1, 2] else
              'Spring' if x in [3, 4, 5] else
              'Summer' if x in [6, 7, 8] else
              'Autumn'
)

print(df[['dt_iso', 'month', 'season']].head())
```

```python
print('summary statistics:')
print(df.describe())
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x = 'season', y= 'temp', palette = 'coolwarm')
plt.title('Temperature Distribution by season')
plt.xlabel("Season")
plt.ylabel('Tempreture (C )')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='pressure', y='temp', hue='season', palette='viridis')
plt.title("Relationship between Pressure and Temperature")
plt.xlabel("Pressure (hPa)")
plt.ylabel("Temperature (°C)")
plt.legend(title="Season")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='season', y='clouds_all', ci=None, palette='Blues_d')
plt.title("Average Cloud Cover by Season")
plt.xlabel("Season")
plt.ylabel("Cloud Cover (%)")
plt.show()
```

```python
strong_wind = df[df['wind_speed'] > 20]
print(f"Number of days with strong winds (>20 m/s): {len(strong_wind)}")
```

```python
extreme_heat = df[df['temp'] > 40]
print(f"Number of days with extreme heat (>40°C): {len(extreme_heat)}")
```

```python
monthly_avg_temp = df.groupby('month')['temp'].mean()

plt.figure(figsize=(10, 6))
monthly_avg_temp.plot(kind='bar', color='orange')
plt.title("Average Monthly Temperature")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.xticks(ticks=np.arange(12), labels=[
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='weather_main', order=df['weather_main'].value_counts().index, palette='viridis')
plt.title("Distribution of Weather Conditions")
plt.xlabel("Count")
plt.ylabel("Weather Condition")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='humidity', y='clouds_all', hue='season')
plt.title("Relationship between Humidity and Cloud Cover")
plt.xlabel("Humidity (%)")
plt.ylabel("Cloud Cover (%)")
plt.legend(title="Season")
plt.show()
```
